```python
class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    def start(self):
        print(self.brand, "car started")

    def stop(self):
        print(self.brand, "car stopped")
```

```python
car1 = Car("Toyota", "Red")
car1.start()
car1.stop()
```

```python
class LoginPage:
    def __init__(self):
        self.correct_username = "admin"
        self.correct_password = "1234"

    def login(self, username, password):
        if username == self.correct_username and password == self.correct_password:
            return "Login Successful"
        return "Invalid credentials"
```

Create a login page to do a headless start

```python
page = LoginPage()

print(page.login("admin", "1234"))
print(page.login("wrong", "wrong"))
```

```python
class Animal:
    def eat(self):
        print("Eating")

    def sleep(self):
        print("Sleeping")

class Dog(Animal):
    def bark(self):
        print("Barking")

d = Dog()
d.eat()
d.sleep()
d.bark()
```

```python
class BasePage:
    def open(self):
        print("Opening browser")

    def close(self):
        print("Closing browser")
```

```python
class LoginPage(BasePage):
    def login(self, username, password):
        print("Logging in with", username)

page = LoginPage()

page.open()
page.login("admin", "1234")
page.close()
```

```python
class Animal:
    def sound(self):
        print("Animal sound")

class Dog(Animal):
    def sound(self):
        super().sound()
        print("Dog bark")

d = Dog()
d.sound()
```
